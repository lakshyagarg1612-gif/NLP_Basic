```python
import pandas as pd
df=pd.read_csv("https://github.com/singhrau0/Big-Data-Preprocessing/raw/refs/heads/main/youtubesample.csv")
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
ans_list=[]
for i in df['description']:
    sorted_list=[]
    word=i.split()
    for j in word:
        if j.startswith('http'):
            word.remove(j)
    clean=" ".join(word)
    ans_list.append(clean)
df['description']=ans_list
df['description']
```

```python
sorted_list=[]
for i in df['description']:
    s=""
    for j in i:
        if j.isalpha() or j==" ":
            s=s+j
    s=s.split()
    cleaned_data=(" ".join(s)).lower()
    sorted_list.append(cleaned_data)
df['description']=sorted_list
```
